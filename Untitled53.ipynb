```python
import os,cv2,pickle
import numpy as np
import pickle
from mtcnn.mtcnn import MTCNN
from keras_facenet import FaceNet
from sklearn.metrics.pairwise import cosine_similarity
from cryptography.fernet import Fernet
```

```python
embedder = FaceNet()
detector = MTCNN()
```

```python
def extract_face(img, size=(160,160)):
    faces = detector.detect_faces(img)
    if not faces:
        return None
    x,y,w,h = faces[0]['box']
    face = img[y:y+h, x:x+w]
    return cv2.resize(face, size)
```

```python
DATASET_PATH="/content/drive/MyDrive/dataset/image_dataset_smaller"
```

```python
key = Fernet.generate_key()
cipher = Fernet(key)

db = {}

for person in os.listdir(DATASET_PATH):
    person_path = os.path.join(DATASET_PATH, person)
    if not os.path.isdir(person_path):
        continue
    for img_name in os.listdir(person_path):
        img = cv2.imread(os.path.join(person_path, img_name))
        face = extract_face(img)
        if face is None:
            continue
        emb = embedder.embeddings([face])[0]
        db.setdefault(person, []).append(emb)

encrypted = cipher.encrypt(pickle.dumps(db))
open("/content/drive/MyDrive/faces.db","wb").write(encrypted)
open("/content/drive/MyDrive/key.key","wb").write(key)
```

```python
db = pickle.loads(cipher.decrypt(open("/content/drive/MyDrive/faces.db","rb").read()))

def recognize(img):
    face = extract_face(img)
    if face is None:
        return "No Face"
    emb = embedder.embeddings([face])[0]
    best, score = "Unknown", 0
    for user, embs in db.items():
        sim = cosine_similarity([emb], embs).max()
        if sim > score and sim > 0.75:
            best, score = user, sim
    return best
```
